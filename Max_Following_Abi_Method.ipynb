```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_190518.txt')
```

```python
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
dmax = df.groupby(['STATION', 'LINENAME', 'UNIT', 'SCP'])[['ENTRIES']].max()
dmax.head(25)
```

```python
dmin = df.groupby(['STATION', 'LINENAME', 'UNIT', 'SCP'])[['ENTRIES']].min()
dmin.head(25)
```

```python
d_diff = dmax - dmin
d_diff.head(25)
```

```python
d_diff.groupby(['STATION', 'LINENAME']).agg({'ENTRIES' : 'sum'}).head()
```

```python
df['DATETIME'] = df['DATE'] + ' ' + df['TIME']
df['DATETIME'] = pd.to_datetime(df['DATETIME'], format = '%m/%d/%Y %H:%M:%S')
```

```python
df.head()
```

```python
volume_table = d_diff.groupby(['STATION', 'LINENAME']).agg({'ENTRIES' : 'sum'})
```
