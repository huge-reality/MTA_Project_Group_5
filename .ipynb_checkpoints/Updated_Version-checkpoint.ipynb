```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_180505.txt')
```

```python
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
# Create DATETIME column and convert string to datetime object

df['DATETIME'] = df['DATE'] + ' ' + df['TIME']
df['DATETIME'] = pd.to_datetime(df['DATETIME'], format="%m/%d/%Y %H:%M:%S")
df.drop(columns=['DATE', 'TIME'], inplace=True)
```

```python
df['DAY'] = df['DATETIME'].dt.day_name()
```

```python
df.head()
```

```python


df['COUNT'] = df.groupby(['UNIT', 'SCP'])['ENTRIES'].transform(lambda x: x.diff())
```

```python
df.head(150)
```

```python
df['COUNT'].describe()
```

```python
df = df[df['COUNT'] >= 0]
```

```python
len(df)
df['COUNT'].max()
```

```python
df.describe()
```

```python
len(df)
```

```python
df = df[df['COUNT'] < 15000]
df.describe()
```

```python
df.head()
```

```python
final = df.groupby(['STATION', 'LINENAME']).agg({'COUNT' : 'sum'})
```

```python
final.reset_index(inplace=True)
final.head()
```

```python
final.sort_values('COUNT', ascending=False)
```
