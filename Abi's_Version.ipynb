```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_190511.txt')
```

```
df.columns = df.columns.str.strip()
```

```
df.head()
```

```
df['STATION'].unique()
```

```
df['STATION']
```

```
#dmax = df.groupby(['STATION', 'LINENAME', 'C/A', 'SCP'])[['ENTRIES']].max()
dmax = df.groupby(['STATION', 'LINENAME', 'UNIT', 'SCP'])[['ENTRIES']].max()
dmax.head(25)
```

```
dmin = df.groupby(['STATION', 'LINENAME', 'UNIT', 'SCP'])[['ENTRIES']].min()
dmin.head(25)
```

```
d_diff = dmax - dmin
d_diff.head(25)
```

```
d_diff.groupby(['STATION', 'LINENAME']).agg({'ENTRIES' : 'sum'}).head()
```
